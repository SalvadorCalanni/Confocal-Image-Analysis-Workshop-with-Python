```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import seaborn as sns
```

```python
ruta = "Imagen.tiff"
imagen = cv2.imread(ruta)
imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB) #Por alguna razn, cv2 lee las imagenes en BGR, as que hay que convertirlas a RGB
plt.imshow(imagen)
```

```python
canal_rojo = imagen[:,:,0] #Canal rojo
canal_verde = imagen[:,:,1] #Canal verde
```

```python
imagen_rojo_bin = (canal_rojo > 80) 
plt.imshow(imagen_rojo_bin)
```

```python
imagen_verde_bin = canal_verde > 50
plt.imshow(imagen_verde_bin)
```

```python
imagen_bin = np.logical_and(imagen_verde_bin, imagen_rojo_bin)
plt.imshow(imagen_bin, cmap ="gray")
```

```python
ruta = "Imagen3.tiff"
imagen = cv2.imread(ruta)
imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB) #Por alguna razn, cv2 lee las imagenes en BGR, as que hay que convertirlas a RGB
plt.imshow(imagen)
```

```python
canal_rojo = imagen[:,:,0] #Canal rojo
canal_verde = imagen[:,:,1] #Canal verde
canal_azul = imagen [:,:,2] #Canal azul
```

```python
imagen_rojo_bin = (canal_rojo > 60) 
plt.imshow(imagen_rojo_bin)
```

```python
imagen_verde_bin = (canal_verde > 80) 
plt.imshow(imagen_verde_bin)
```

```python
imagen_azul_bin = (canal_azul > 60) 
plt.imshow(imagen_azul_bin)
```

```python
imagen_bin_rojo = np.logical_and(imagen_rojo_bin, imagen_azul_bin)
imagen_bin_verde = np.logical_and(imagen_verde_bin, imagen_azul_bin)
imagen_bin = np.logical_and(imagen_verde_bin, imagen_rojo_bin, imagen_azul_bin)
plt.imshow(imagen_bin, cmap ="gray")
```

```python
def contar_puntos(imagen_bin):
    contornos, _ = cv2.findContours(imagen_bin, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    return len(contornos)
```

```python
imagen_bin_rojo = (imagen_bin_rojo * 255).astype('uint8')
imagen_bin_verde = (imagen_bin_verde * 255).astype('uint8')
imagen_bin_uint8 = (imagen_bin * 255).astype('uint8')
```

```python
puntos_rojo = contar_puntos(imagen_bin_rojo)
puntos_verde = contar_puntos(imagen_bin_verde)
puntos_combinados = contar_puntos(imagen_bin_uint8)

print(f"Puntos en imagen_rojo: {puntos_rojo}")
print(f"Puntos en imagen_verde: {puntos_verde}")
print(f"Puntos en la imagen combinada: {puntos_combinados}")
```
