```python
# Vamos a usar matplotlib para graficar. 
# Ejemplo: Un gráfico muy sencillo con matplotlib
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
```

```python
#SI queremos, podemos seguir agregando todo lo que queramos al gráfico:

#Este es el mismo gráfico de antes:
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)

# Ahora le metemos otra serie de números:

y2 = np.cos(x)

plt.plot(x, y2)

#Ahora un punto cualquiera:
plt.scatter(2, 0) 
```

```python
#Otra cosa que puede ser interesante, es que cada cosa graficada tenga una leyenda.
#Para eso vamos agregando los elementos al gráfico con plt.plot() y plt.scatter() y luego llamamos a plt.legend()
#Este es el mismo gráfico de antes:
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='seno')

y2 = np.cos(x)

plt.plot(x, y2, label='coseno')

plt.scatter(2, 0 , label='punto') 

plt.legend() #Esto arma un cuadrito sobre el gráfico con todas las labels que fuimos armando
```

```python
#Si queremos que las cosas se ploteen en gráficos distintos, tenemos que avisarle que "cierre" el gráfico anterior:
#Este es el mismo gráfico de antes:
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)

plt.show() #Con este comando le decimos que muestre el gráfico anterior y que empiece uno nuevo


y2 = np.cos(x)

plt.plot(x, y2)

#Ahora un punto cualquiera:
plt.scatter(2, 0) 
```

Graficando imágenes

Una imagen en escala de grises puede pensarse como un array bidimensional en  donde cada número indica la intensidad de gris:





Se usan codificaciones que ocupen 1 byte por número (pueden ser decimales entre 0  y 1, o enteros entre 0 y 255).

```python
#Creamos un array con numeros enteros aleatorios en numpy

arr = np.random.randint(0,256,(10,12))
```

```python
arr.shape
```

```python
arr
```

¡Qué bárbaro!

```python
#La librería matplotlib tiene una función para mostrar una array con numeros en colores. O sea, una imagen.
plt.imshow(arr)
```

Por defecto, acá usó un mapa de colores que va desde el violeta para los números más bajos al amarillo para los más altos.
Hay un parámetro que le podemos dar a esa función con el mapa de colores que queramos. Se llama cmap y pueden ve los valores posibles acá: https://matplotlib.org/stable/users/explain/colors/colormaps.html

```python
plt.imshow(arr, cmap='gray')
```

Ahora con una imagen real

```python
ruta = 'Fotos\\retina.tif'
```

```python
#guardamos la imagen en un array
retina = plt.imread(ruta)

retina.shape

```

En este caso tenemos un array tridimensional, de tamaño n×m×3, donde n y m son la cantidad de filas y columnas (en pixeles) y 3 son las 3 capas de color.



```python
plt.imshow(retina)
```

```python
#separamos la imagen en sus 3 canales
r = retina[:,:,0]
g = retina[:,:,1]
b = retina[:,:,2]
```

```python
#mostrar los distintos canales por separado, por ejemplo: 
plt.imshow(g, cmap='gray')
```

```python
intensidad_g = np.mean(g, axis=1)
intensidad_r = np.mean(r, axis=1)
intensidad_b = np.mean(b, axis=1)
```

```python

plt.plot(intensidad_g, color='green')
plt.plot(intensidad_r, color='red')
plt.plot(intensidad_b, color='blue')

plt.ylabel('Intensidad')
plt.xlabel('px')
```
